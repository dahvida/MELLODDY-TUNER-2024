### Get necessary imports

```python
from molfeat.trans.pretrained import PretrainedDGLTransformer
from rdkit import Chem
import numpy as np
```

### Helper function to parse the output

```python
def parse_gin_output(gin_output):
    out = []
    for i in range(gin_output.shape[1]):
        out.append(gin_output[0,i])
    return tuple(out)
```

### Make test molecule

```python
smi = "C1=CN=CN1"
```

### Get encoder

```python
gin_infomax = PretrainedDGLTransformer(kind='gin_supervised_infomax', dtype=float)
```

### Run encoder, check outputs

```python
output = gin_infomax(smi)
parsed_output = parse_gin_output(output)
```
